```
import kagglehub, os

datasets = {
    'Logo':      'siddharthkumarsah/logo-dataset-2341-classes-and-167140-images',
    'Font':      'muhammadardiputra/font-recognition-data',
}

paths = {}
for name, slug in datasets.items():
    print(f'📥 Downloading {name} dataset...')
    try:
        path = kagglehub.dataset_download(slug)
        paths[name] = path
        print(f'✅ {name}: {path}')
    except Exception as e:
        print(f'⚠️  {name}: {e}')
```

```
import pandas as pd
import numpy as np

# Generate synthetic Slogan, Startups, Marketing datasets for inspection
np.random.seed(42)

# Slogan dataset
df_slogan = pd.DataFrame({
    'slogan': ['Just Do It', 'Think Different', 'Because You Are Worth It', 'Open Happiness', 'Impossible is Nothing'],
    'brand': ['Nike', 'Apple', "L'Oreal", 'Coca-Cola', 'Adidas'],
    'industry': ['Fashion', 'Tech', 'Beauty', 'Food', 'Fashion'],
    'tone': ['bold', 'minimalist', 'elegant', 'vibrant', 'bold'],
    'word_count': [3, 2, 5, 2, 3]
})

# Startups dataset
df_startups = pd.DataFrame({
    'company': ['TechNova', 'FreshBite', 'LuxeWear', 'EduSpark', 'BuildRight'],
    'industry': ['Tech / Software', 'Food & Beverage', 'Fashion / Apparel', 'Education', 'Manufacturing'],
    'region': ['North America', 'Europe', 'Asia Pacific', 'South Asia', 'Latin America'],
    'target_audience': ['Millennials', 'Gen Z', 'Professionals', 'Students', 'Businesses'],
    'personality': ['minimalist', 'vibrant', 'luxury', 'bold', 'minimalist']
})

# Marketing dataset
df_marketing = pd.DataFrame({
    'platform': ['Instagram', 'Facebook', 'Twitter/X', 'Instagram', 'Facebook'],
    'region': ['North America', 'Europe', 'Asia Pacific', 'Middle East', 'Latin America'],
    'objective': ['Brand Awareness', 'Engagement', 'Conversion', 'Brand Awareness', 'Engagement'],
    'ctr': [3.2, 1.8, 2.1, 2.9, 1.6],
    'roi': [145, 178, 312, 134, 198],
    'engagement': [68, 52, 61, 74, 49]
})

print('📋 Slogan Dataset:')
print(df_slogan.head())
print(df_slogan.info())
print('\n📋 Startups Dataset:')
print(df_startups.head())
print(df_startups.info())
print('\n📋 Marketing Dataset:')
print(df_marketing.head())
print(df_marketing.info())
```

```
print('🔍 Data Quality Checks\n')
for name, df in [('Slogan', df_slogan), ('Startups', df_startups), ('Marketing', df_marketing)]:
    print(f'--- {name} Dataset ---')
    print(f'Shape: {df.shape}')
    print(f'Missing values:\n{df.isnull().sum()}')
    print(f'Duplicates: {df.duplicated().sum()}\n')
```

```
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('BrandSphere AI — Dataset Distributions', fontsize=15, fontweight='bold')

# Industry distribution
sns.countplot(data=df_startups, x='industry', ax=axes[0], palette='YlOrBr')
axes[0].set_title('Startup Industries')
axes[0].tick_params(axis='x', rotation=30)

# Slogan word frequency
from collections import Counter
all_words = ' '.join(df_slogan['slogan']).lower().split()
word_freq = Counter(all_words).most_common(10)
axes[1].barh([w[0] for w in word_freq], [w[1] for w in word_freq], color='#1B3A6B')
axes[1].set_title('Slogan Word Frequency')
axes[1].invert_yaxis()

# Platform CTR comparison
sns.barplot(data=df_marketing, x='platform', y='ctr', ax=axes[2], palette='coolwarm')
axes[2].set_title('CTR by Platform')

plt.tight_layout()
plt.savefig('week1_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Visualizations saved')
```

```
# Display sample logos using OpenCV + matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt

if 'Logo' in paths:
    logo_root = paths['Logo']
    sample_imgs = []
    sample_names = []
    for root, dirs, files in os.walk(logo_root):
        for f in files:
            if f.lower().endswith(('.jpg','.jpeg','.png')):
                sample_imgs.append(os.path.join(root, f))
                sample_names.append(os.path.basename(root))
            if len(sample_imgs) == 8:
                break
        if len(sample_imgs) == 8:
            break

    fig, axes = plt.subplots(2, 4, figsize=(16, 8))
    fig.suptitle('Sample Logos from Dataset', fontsize=14, fontweight='bold')
    for ax, (path, name) in zip(axes.flat, zip(sample_imgs, sample_names)):
        img = cv2.imread(path)
        img = cv2.cvtColor(cv2.resize(img, (128,128)), cv2.COLOR_BGR2RGB)
        ax.imshow(img)
        ax.set_title(name[:12], fontsize=9)
        ax.axis('off')
    plt.tight_layout()
    plt.savefig('week1_sample_logos.png', dpi=150, bbox_inches='tight')
    plt.show()
    print('✅ Sample logos displayed')
```
