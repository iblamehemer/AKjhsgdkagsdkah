```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime, timedelta
import random

np.random.seed(42)
random.seed(42)

# Simulate collected feedback data
modules = ['Logo & Design Studio', 'Creative Content Hub', 'Campaign Studio', 'Brand Aesthetics Engine']
comments_positive = ['Great logo!', 'Love the colors', 'Very professional', 'Excellent quality', 'Perfect for my brand']
comments_negative = ['Too generic', 'Colors too dark', 'Needs more variety', 'Tagline too short', 'Improve fonts']
comments_neutral  = ['Decent output', 'Works fine', 'Average quality', 'Meets expectations', 'Acceptable result']

feedback_data = []
for i in range(150):
    rating = random.choices([1,2,3,4,5], weights=[5,10,20,35,30])[0]
    module = random.choice(modules)
    if rating >= 4:
        comment = random.choice(comments_positive)
        sentiment = 'positive'
    elif rating <= 2:
        comment = random.choice(comments_negative)
        sentiment = 'negative'
    else:
        comment = random.choice(comments_neutral)
        sentiment = 'neutral'
    feedback_data.append({
        'timestamp': (datetime.now() - timedelta(days=random.randint(0,30))).isoformat(),
        'module': module, 'rating': rating,
        'comment': comment, 'sentiment': sentiment,
        'session_id': f'session_{i:04d}'
    })

df_fb = pd.DataFrame(feedback_data)
print(f'✅ Feedback dataset: {df_fb.shape}')
print(df_fb.head())
print(f'\n📊 Rating distribution:\n{df_fb["rating"].value_counts().sort_index()}')
```

```
# Save feedback to CSV (simulating Google Drive storage)
df_fb.to_csv('feedback_data.csv', index=False)
print('✅ Feedback saved to feedback_data.csv')
```

```
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('BrandSphere AI — Feedback Intelligence Dashboard', fontsize=15, fontweight='bold')

# Rating distribution
rating_counts = df_fb['rating'].value_counts().sort_index()
axes[0][0].bar(rating_counts.index, rating_counts.values,
               color=['#E05A5A','#F4A261','#C9A84C','#3ECFB2','#2D6A4F'])
axes[0][0].set_title('Rating Distribution (1-5)', fontweight='bold')
axes[0][0].set_xlabel('Rating'); axes[0][0].set_ylabel('Count')

# Ratings by module
module_avg = df_fb.groupby('module')['rating'].mean().sort_values()
axes[0][1].barh(module_avg.index, module_avg.values, color='#1B3A6B')
axes[0][1].set_title('Average Rating by Module', fontweight='bold')
axes[0][1].set_xlabel('Average Rating')
for i, v in enumerate(module_avg.values):
    axes[0][1].text(v+0.02, i, f'{v:.2f}', va='center', fontsize=9)

# Sentiment distribution
sent_counts = df_fb['sentiment'].value_counts()
axes[1][0].pie(sent_counts.values, labels=sent_counts.index,
               colors=['#3ECFB2','#C9A84C','#E05A5A'],
               autopct='%1.1f%%', startangle=90)
axes[1][0].set_title('Sentiment Distribution', fontweight='bold')

# Ratings over time
df_fb['date'] = pd.to_datetime(df_fb['timestamp']).dt.date
daily_avg = df_fb.groupby('date')['rating'].mean()
axes[1][1].plot(daily_avg.index, daily_avg.values, color='#C9A84C', linewidth=2, marker='o', markersize=4)
axes[1][1].set_title('Daily Average Rating Trend', fontweight='bold')
axes[1][1].set_xlabel('Date'); axes[1][1].set_ylabel('Avg Rating')
axes[1][1].tick_params(axis='x', rotation=30)
axes[1][1].axhline(y=df_fb['rating'].mean(), color='#E05A5A', linestyle='--', alpha=0.5, label='Overall avg')
axes[1][1].legend()

plt.tight_layout()
plt.savefig('feedback_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Feedback dashboard saved')
```

```
# NLP analysis of feedback comments
import nltk
from collections import Counter
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

def get_top_words(df_subset, n=10):
    all_words = []
    for comment in df_subset['comment']:
        tokens = word_tokenize(comment.lower())
        tokens = [t for t in tokens if t.isalpha() and t not in stop_words and len(t) > 2]
        all_words.extend(tokens)
    return Counter(all_words).most_common(n)

print('📋 Top words in POSITIVE feedback:')
print(get_top_words(df_fb[df_fb['sentiment']=='positive']))
print('\n📋 Top words in NEGATIVE feedback:')
print(get_top_words(df_fb[df_fb['sentiment']=='negative']))
```

```
# Model refinement simulation
print('🔄 Model Refinement based on Feedback\n')

# Identify lowest rated module
lowest_module = df_fb.groupby('module')['rating'].mean().idxmin()
lowest_avg = df_fb.groupby('module')['rating'].mean().min()
print(f'📉 Lowest rated module: {lowest_module} ({lowest_avg:.2f}/5)')

# Simulate weight adjustment
print('\n🔧 Adjusting model weights based on feedback...')
refinements = {
    'Logo & Design Studio': 'Increase style diversity weight by 15%',
    'Creative Content Hub': 'Fine-tune tagline length constraint (8-12 words)',
    'Campaign Studio': 'Boost regional personalization factor by 20%',
    'Brand Aesthetics Engine': 'Recalibrate color harmony scoring threshold',
}
for module, action in refinements.items():
    avg = df_fb[df_fb['module']==module]['rating'].mean()
    priority = 'HIGH' if avg < 3.5 else 'MED' if avg < 4.0 else 'LOW'
    print(f'  [{priority}] {module}: {action}')

print('\n✅ Refinement plan generated based on aggregated user feedback')
```
